```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
```

```python
y_and = np.array([0, 0, 0, 1])
model_and = Perceptron(max_iter=1000, eta0=0.1)
model_and.fit(X, y_and)
print("Predictions for AND:", model_and.predict(X))
```

```python
y_or = np.array([0, 1, 1, 1])
model_or = Perceptron(max_iter=1000, eta0=0.1)
model_or.fit(X, y_or)
print("Predictions for OR:", model_or.predict(X))
```

```python
y_xor = np.array([0, 1, 1, 0])
model_xor = Perceptron(max_iter=1000, eta0=0.1)
model_xor.fit(X, y_xor)
print("Predictions for XOR:", model_xor.predict(X))
```

```python
def plot_boundary(ax, X, y, model, title):
    xx, yy = np.meshgrid(
        np.linspace(-0.5, 1.5, 200),
        np.linspace(-0.5, 1.5, 200)
    )
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=100)

    ax.set_title(title)
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

plot_boundary(axes[0], X, y_and, model_and, "AND Gate")
plot_boundary(axes[1], X, y_or,  model_or,  "OR Gate")
plot_boundary(axes[2], X, y_xor, model_xor, "XOR Gate")

plt.tight_layout()
plt.show()
```
